## On uniformly distributed data

```python
%load_ext Cython
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'Define some random pts in space'
L = 1000
n_pts = 500
xs, ys = np.random.randint(low=0, high=L, size=(2, n_pts))
plt.scatter(xs, ys);
```

```python
'Compute distance matrix between all pts'
D = np.zeros((n_pts, n_pts))
for i in range(n_pts):
    for j in range(n_pts):
        D[i, j] = np.linalg.norm((xs[i]-xs[j], ys[i]-ys[j]))
import seaborn as sns
sns.heatmap(D);
```

```python
from scipy.spatial.distance import cdist
crd = np.hstack((xs[:,np.newaxis],ys[:,np.newaxis]))
D = cdist(crd,crd)
```

- Cool [link](https://pro.arcgis.com/en/pro-app/tool-reference/spatial-statistics/h-how-multi-distance-spatial-cluster-analysis-ripl.htm) instead.

```python
def K_t_1(t):
    """
    Univariate Ripley function. From:
    https://en.wikipedia.org/wiki/Spatial_descriptive_statistics
    
    """
    s = 0
    for i in range(n_pts):
        for j in range(n_pts):
            if i != j:
                s += int(D[i, j] < t)
    s *= (L**2)
    s /= (n_pts**2)
    return s

def L_t(t):
    return np.sqrt(K_t(t)/np.pi)
```

```python
def K_t_2(t):
    s  = 0
    for i in range(n_pts-1):
        for j in range(i+1,n_pts):
            s += int(D[i,j] < t)
    s *= (L**2)
    s /= (n_pts**2)
    return s*2
            
```

```python
%%cython
from numpy cimport ndarray as ar

def K_t_3(double t,
         ar[double,ndim =2] W,
         double L,
         int n_pts):
    
    cdef double s  = 0.0
    cdef int i,j
    
    for i in range(n_pts-1):
        for j in range(i+1,n_pts):
            s += int(W[i,j] < t)
            
    s = s * (L**2)
    s = s / float(n_pts)**2
    
    return s*2.0
```

- Ripley on uniform data should follow the prediction.
- As the search range gets close to L there are finite size effects.
- This effects manifests as "disperse data" (_ie_ smaller than prediction).
- Disperse because less "dense" than the uniform prediction (because of the border).
