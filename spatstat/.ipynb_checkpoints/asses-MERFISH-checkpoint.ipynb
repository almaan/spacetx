## Data Structures

We define some custom classes to make the analysis easier, and the remaining code agnostic to whatever data we want to use.

- `DataSet`: abstract base class designed to hold spatial data
- `MerfishData` : child class inheriting from `DataSet`, designed specifically for MERFISH data 

Functions and analysis will be written as to be compatible with the `DataSet` class.

[1]:https://github.com/spacetx-spacejam/tommaso-team/blob/master/mapping_results/merfish_new_seg.csv

```python
class DataSet(ABC):
    """Class to hold expression data"""
    def __init__(self,
                 pth : str,
                )->None:
        
        
        # store original data
        self._data : pd.DataFrame = self._read_data(pth)
        
        # number of data points
        self.N = self._data.shape[0]
        
        # width and height of region of study
        # when applicable
        self.width : Optional[float] = None
        self.height : Optional[float] = None
        
        # generate distance matrix
        self.D : np.ndarray = self._get_weights()
        
        # get ranges of data region
        self.ranges : Dict[str,np.ndarray] = self._get_ranges()
        # compute are for data region
        self.area : float = self._get_area()
        
        # numeric versions of labels
        self.label_to_num : Optional[LE] = None
        
        # unique cell types present in data
        self.uni_types : np.ndarray = np.unique(self.types)
        
            
    @abstractmethod
    def _read_data(self,
                   pth : str,
                  )->pd.DataFrame:
        """read data from path"""
        pass
        
    @abstractproperty
    def crd(self,)->np.ndarray:
        """get coordinates"""
        pass
    
    @abstractmethod
    def _get_area(self,
                  *args,
                  **kwargs,
                 )->float:
        """compute area of data region"""
        pass
    
    @abstractmethod
    def _get_ranges(self,
                  )->Dict[str,np.ndarray]:
        
        """compute ranges of data region"""

        pass
    
    @abstractproperty
    def types(self,)->np.ndarray:
        """vector with cell type of each data point"""
        pass
    
    @property
    def numeric_types(self,
                     )->np.ndarray:
        """numeric version of cell type labels"""
        if self.label_to_num is None:
            self.label_to_num = LE().fit(self.types)
        return self.label_to_num.transform(self.types)
    
    @abstractmethod
    def sample_crd(self,
                   npts : int,
                  )->np.ndarray:
        """sample coordinates from data region"""
        pass
    
    def _get_weights(self,
                    )->np.ndarray:
        """compute distance weights"""
        
        return cdist(self.crd,
                     self.crd,
                     metric = "euclidean")
    
    
    def simulate(self,
                 npts : int,
                )->np.ndarray:
        """simulate CSR in the data region"""
        
        npts = np.random.poisson(npts)
        crd = sample_crd(npts)
        
    def __getitem__(self,x)->pd.DataFrame:
        return self._data.iloc[x,:]
    
    def __len__(self,)->int:
        return self.N
        
     
```

```python
# Define child of DataSet class to hold MERFISH data

class MerfishData(DataSet):
    def __init__(self,
                 data : pd.DataFrame,
                 type_colname : str = "types",
                )->None:
        
        super().__init__(data)
    
        self.type_colname = "types"
        
        
    def _read_data(self,
                   pth : str,
                  )->pd.DataFrame:
        
        return pd.read_csv(pth,
                          sep = ',',
                          header = 0,
                          index_col = 0,
                          )
        
    @property
    def crd(self,)->np.ndarray:
        return self._data[["x","y"]].values
    
    @property
    def types(self,
              )->np.ndarray:
        return self._data["types"].values
    
    def _get_ranges(self,
                   )->Dict[str,np.ndarray]:
        
            
        mxs = self.crd.max(axis =0)
        mns = self.crd.min(axis =0)
        
        _ranges = dict(x = np.array((mns[0],mxs[0])),
                       y = np.array((mns[1],mxs[1])),
                      )
        
        self.width = mxs[0] - mns[0]
        self.height = mxs[1] - mns[0]
                
        return _ranges
    
    def _get_area(self,
                 )->float:
        
        dx = self.ranges["x"][1]-self.ranges["x"][0]
        dy = self.ranges["y"][1]-self.ranges["y"][0]
        
        return dx*dy
    
    def sample_crd(self,
                  npts: int,
                  )->np.ndarray:
        
        xs = np.random.uniform(low = self.ranges["x"][0],
                               high = self.ranges["x"][1],
                               size = npts,
                              ).reshape(-1,1)
        
        ys = np.random.uniform(low = self.ranges["x"][0],
                               high = self.ranges["x"][1],
                               size = npts,
                              ).reshape(-1,1)
        
        return np.hstack((xs,ys))
    
        
```

# Ripley's K

We use the definition of Ripley's K given [here](1) as:
\begin{equation}
K(r) = \frac{A}{n^2}\sum_{i=1}^n\sum_{j=1}^n I_r(x_i,x_j)
\end{equation}

where:
\begin{equation}
I_r(i,j) = \Bigg{\{}
\begin{array}{ll}
1 & ||x_i - x_j|| < r \\
0 & \textrm{otherwise}
\end{array}
\end{equation}

Implementation in `Cython` for efficiency reasons


[1]:https://en.wikipedia.org/wiki/Spatial_descriptive_statistics

```python
%%cython
# Define Ripley's K function

from numpy cimport ndarray as ar

def K_t(double t,
        ar[double,ndim =2] D,
        double A,
        ):
    
    """Ripley's K
    
    Parameters:
    -----------
    
    t : double
        distance to compute metric for
    D : ndarray[double,ndim=2]
        distance matrix
    A : double
        area of region
    
    Returns:
    --------
    Ripley's K at distance t
    
    """
    
    cdef double s  = 0.0
    cdef int i,j
    cdef int n_pts = D.shape[0]
    
    for i in range(n_pts-1):
        for j in range(i+1,n_pts):
            s += int(D[i,j] < t)
            
    s = s * A
    s = s / float(n_pts)**2
    
    return s*2.0
    
```

# PCF - Pair Correlation Function

Using the original definition from [Stoyan and Stoyan](1) of the PCF:

<img src="img/estim_pcf.png">

They describe it for a univariate case, hence the divisor $\hat{\lambda}^2 = (\frac{N_{tot}^2}{A^2})$. For the bivariate case, I've used $\lambda_1\lambda_2 = \frac{n_1}{A}\frac{n_2}{A}$ as the divisor instead (in-line with [this](2) and [this](3))

Hence we get:

\begin{equation}
PCF(r) = \frac{\hat{\rho}(r)}{\lambda_1\lambda_2} = \frac{1}{2\pi r}\sum_{i \in N_1}\sum_{j \in N_2} \frac{k_h(r-||x_i -x_j||)}{A}\times\frac{A^2}{n_1 n_2} = \frac{A}{2\pi r}\sum_{i \in N_1}\sum_{j \in N_2} \frac{k_h(r-||x_i -x_j||)}{n_1 n_2}
\end{equation}

Where $k_h$ is the Epanechnikov Kernel ($e_h$) defined as:

<img src="img/kernel.png">


$h$ is commonly referred to as bandwidth

[1]:https://onlinelibrary.wiley.com/doi/10.1002/zamm.19950750815
[2]:https://link.springer.com/content/pdf/10.1007/s11258-006-9198-0.pdf
[3]:https://wiki.landscapetoolbox.org/doku.php/spatial_analysis_methods:ripley_s_k_and_pair_correlation_function

```python
%%cython
from numpy cimport ndarray as ar
from numpy cimport abs as abs

cimport cython

@cython.boundscheck(False)
@cython.wraparound(False)
def PCF_r(double r,
          ar[double,ndim =2] D,
          double A,
          double h,
          )->float:
    
    """Pair Correlation Function
    
    Parameters:
    ----------
    
    r : double
        radius of annuli
    D : ndarray[double,ndim=2]
        distance matrix
    A : double
        area of region of interest
    h : double
        bandwidth
        
    """
    
    def eh(double t)->float:
        cdef double kh
        kh =  3.0 / 4.0 / h * (1.0-(t/h)**2)
        return kh
    
    cdef double pi = 3.141592653589793
    cdef int n1 = D.shape[0]
    cdef int n2 = D.shape[1]
    cdef double s = 0    
    cdef double dr
    cdef int i,j
    
    for i in range(n1):
        for j in range(n2):
            dr = abs(D[i,j] - r)
            s += (eh(dr) if dr <= h else 0)
            
    s *= (A / (float(n1) * float(n2) * 2 * r * pi))
    return s

```

## Synthetic data

We can generate some synthetic data to check that the `PCF_r` function works as expected. A simple scheme is used and implemented in `generate_pcf_synth`. In short, this:

1. Generates `n_seeds` seeding points
2. Generates type 1 cells at a distance $d \in [r - \delta, r + \delta]$ from the seeding points
3. Generates type 2 cells at a distance $d \in [r - \delta, r + \delta]$ from the type 1 cells

$r$ can either be one value, or sampled from a set of two different values (`radius_1`and `radius_2`). $\delta$ is constant. 

_NOTE_ : maybe not the most realistic data, but ok as a sanity check


```python
# Define function to generate synthetic data

def generate_pcf_synth(n_seeds : int,
                      radius_1 : float = 5.0,
                      radius_2 : Optional[float] = None,
                      delta :  float = 1.0,
                      max_paired : int = 10,
                      crd_low : float = 0,
                      crd_high : float = 400,
                      seed : Optional[int] = None,
                     )->Tuple[np.ndarray,np.ndarray,np.ndarray,float]:
    
    # radii sampling function
    sample_r = lambda r : np.random.uniform(r-delta,
                                            r+delta)
    
    # if seeding should be used
    if seed is not None:
        np.random.seed(seed)
    
    # if 1 or 2 radii values should be used
    if radius_2 is None:
        radius_2 = radius_1
    
    # compute area
    area = (crd_high-crd_low)**2

    # generate seeding set
    crd_seed = np.random.uniform(low  = crd_low,
                                 high = crd_high,
                                 size = (n_seeds,2))

    # dict to hold coordinates
    crd_list = dict(seed = crd_seed,
                    crd_1 = [],
                    crd_2 = [])

    # generate type 1 and type 2 cells
    for crd,ref,mp in zip(["crd_1","crd_2"],
                          ["seed","crd_1"],
                          [max_paired,1],
                         ):
        
        # iterate over reference set
        for p in range(len(crd_list[ref])):
            
            # get reference cell position
            center = crd_list[ref][p,:] 
            
            # to determine which radii value
            # that should be used
            f = np.random.binomial(1,0.5)
            
            # sample number of paired cells
            # for the reference cell
            if mp > 1:
                n_paired = int(np.random.uniform(1,max_paired))
            else:
                n_paired = 1
            
            # generate paired cells
            for n in range(n_paired):
                
                # sample radii value
                r = sample_r((radius_1 if f > 0 else radius_2))
                
                # sample angle
                theta = np.random.uniform(-np.pi,np.pi)
                
                # generate displacenemt vector
                _xy = np.array((r*np.cos(theta),r*np.sin(theta)))
                # get position for paired cell
                _xy = center + _xy
                # store positition of cell
                crd_list[crd].append(_xy)
        # convert to array
        crd_list[crd] = np.array(crd_list[crd])

    # extract values
    crd_1 = crd_list["crd_1"]
    crd_2 = crd_list["crd_2"]
    crd_2 = np.vstack((crd_2,crd_seed))

    
    return (crd_1,crd_2,crd_seed,area)
```

We generate some synthetic data and then visualize the result

```python
radius_1,radius_2, delta = 4,10,1 

crd_1,crd_2,crd_seed,area = generate_pcf_synth(n_seeds=20,
                                              radius_1 = radius_1,
                                              radius_2 = radius_2,
                                              delta = delta,
                                              max_paired = 20,
                                              seed = 1337)
```

```python
# visualize synhtetic data
fig,ax = plt.subplots(1,1, figsize = (10,10))

ax.scatter(crd_2[:,0],
           crd_2[:,1],
           alpha = 0.3,
           label = "Cell type 1")

ax.scatter(crd_1[:,0],
           crd_1[:,1],
           alpha =0.3,
           label = "Cell type 2",
          )

ax.set_title("Cisualization synthetic data")
ax.legend()

plt.show()
```

Compute the `PCF` over a set value of distances - here from zero to width/height of the region.

```python
t_dists = np.arange(0.1,
                    100,
                    delta,
                   )

t_D = cdist(crd_1,
            crd_2)

t_pcf = [PCF_r(r = r,
               D = t_D, 
               A = area,
               h = delta) for r in t_dists]
```

Visualize the result, the values for `radius_1` and `radius_2` (if specified) will be plotted as dashed vertical lines as references.

```python
fig,ax = plt.subplots(1,1, figsize = (20,5))

ax.plot(t_dists,
        t_pcf,
        "-o",
        color = "black",
        markerfacecolor = "red",
        markeredgecolor = "black",
        label = "data",
       )

ax.axvline(x = radius_1,
            color = "blue",
            alpha = 0.9,
            linestyle = "dashed",
            label = "r = {}".format(radius_1),
            )

if radius_2 is not None:
    ax.axvline(x = radius_2,
                color = "green",
                alpha = 0.9,
                linestyle = "dashed",
                label = "r = {}".format(radius_2),
                )

ax.axhline(y = 1,
            color = "black",
            alpha = 0.9,
            label = "theoretical"
           )

ax.legend()
ax.set_title("PCF function evaluation | Bandwidth {}".format(delta))

plt.show()
```

**COMMENTS**

- The granularity of the distance mesh impacts the results quite a lot, relatively fine grids should be used
- The bandwidth also has an impact on the results, with synthetic data the choice is easy but should be evaluated in real data. Stoyan and Stoyan give a rule of thumb (p.285).
